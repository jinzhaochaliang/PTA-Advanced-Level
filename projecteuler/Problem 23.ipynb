```python
def isguoshen(num):
    n=0
    for i in range(1,num):
        if num%i==0:
            n+=i
    if n>num:
        return True
    else:
        return False
```

```python
alist=[]
for i in range(1,28111):
    if isguoshen(i):
        alist.append(i)
```

```python
len(alist)
```

```python
ssum={0}
for i in range(6961):
    for j in range(i,6961):
        ssum.add(alist[i]+alist[j])
ssum
```

```python
tsum=[]
for i in range(1,28123):
    if i not in ssum:
        tsum.append(i)
sum(tsum)
```

```python
import math

def Full(x):
    list1 = [1]
    for i in range(2,int(math.sqrt(x)+1)):
        if x % i == 0:
            list1.append(i)
            if x/i not in list1:
                list1.append(x/i)
                
    if sum(list1) > x:
        return True
    else:
        return False      
list1 = []
for i in range(1,28111):
    if Full(i):
        list1.append(i)
list2 = {0}
for i in range(6961):
    for n in range(i,6961):
        list2.add(list1[i]+list1[n])
list3=[]
for i in range(1,28123):
    if i not in list2:
        list3.append(i)

print(sum(list3))
```
