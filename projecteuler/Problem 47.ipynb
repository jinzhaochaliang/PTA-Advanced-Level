```python
def isprime(n):
    if n>1:
        if n == 2:
            return True
        if n % 2 ==0:
            return False
        for i in range(3,int(n**0.5+1),2):
            if n % i ==0:
                return False
        return True
    return False
def find(num):
    flist = []
    for i in range(2,int(num**0.5+1)):
        if isprime(i):
            if num%i==0:
                flist.append(i)
                num1=int(num/i)
                if isprime(num1):
                    flist.append(num1)
                else:
                    flist=flist+find(num1)
    return list(set(flist))
```
