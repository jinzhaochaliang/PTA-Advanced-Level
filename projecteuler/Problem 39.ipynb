```python
solve=[]
for a in range(1,250):
    for b in range(a,500-a):
        for c in range(b,500):
            if a*a+b*b==c*c:
                p=a+b+c
                solve.append([p,a,b,c])
cal=[]
master={}
for each in solve:
    if each[0] not in cal:
        cal.append(each[0])
        master[each[0]]=1
    else:
        master[each[0]]+=1
print(master)
```
