```python
def sumprime(n):
    count=0
    t=2
    t4=0
    i=1
    while(i<n):
        i+=t
        flag=True
        for j in range(2,int(i**0.5+1)):
            if i%j==0:
                flag=False
                break
        if flag:
            count+=1
        t4+=1
        if t4%4==0:
            t+=2
        print(count/(t4+1))
        if count/(t4+1)<0.1:
            break
    return i
```
